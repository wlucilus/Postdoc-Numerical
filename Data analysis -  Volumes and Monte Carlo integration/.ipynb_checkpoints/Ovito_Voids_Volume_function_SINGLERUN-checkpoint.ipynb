# Calculation of the void space between water molecules solvating a SCN molecule.
### This is a modular file, that calculates separately (or in a singlerun, if needed) 
### the voids between the first solvation shell of some molecule, generate a xyz void file,
### then integrate the void file to find the volume using a monte carlo integration (see convergence at the end of file). 

Carefull with:
____
+ selectCarbonNNN , uses carbon id, which varies with MDinputFile.xyz ( CarbonType  MD=2 Aimd=3) , also 8 oxygens in the first solvation needs the proper id
+ DefineVdw , need the proper particle name to define the types
    - we are configure so that SCN molecules have zero size, so their vdw does not subtract from the void volume

+ WriteFile: Also atomType id dependent

+ ParticleVis.radius, is NOT the vdW radii of the particles
    + use ComputeProperty to setup radii
    
    
Also check the code Ovito_voids_volume_mulprocessing.py this is a parallel version of this code. (cannot be run in jupyter notebooks)

# This function will calculates the void after the Voronoi tesselation is done
## will generate a file containing the molecules and the voids in a given radii (shell)

```python

def calculateVoidsScy(frame, data):

    # This user-defined modifier function gets automatically called by OVITO whenever the data pipeline is newly computed.
    # It receives two arguments from the pipeline system:
    #
    #    frame - The current animation frame number at which the pipeline is being evaluated.
    #    data  - The DataCollection passed in from the pipeline system.
    #            The function may modify the data stored in this DataCollection as needed.


    voronoiVertices = data.surfaces['voronoi-polyhedra'].get_vertices()
    atomPos = data.particles['Position']
    atomType = data.particles['Particle Type']
    verticesVoidsx = np.zeros(3*len(voronoiVertices))
    verticesVoidsy = np.zeros(3*len(voronoiVertices))
    verticesVoidsz = np.zeros(3*len(voronoiVertices))
    cavityVoids = np.zeros(3*len(voronoiVertices))
    maxD = max(farthest)
    k = -1

    start = time.time()

    #print('List from scypY:')
    for i,atom in enumerate(atomPos):
        if len(atomPos) > 500:
            print('Too many particles to calculate. Select less particles (use ovito delete selected) or increase the limit via code.')
            break

        cavityT = data.particles['Cavity Radius'][i]

        #monster code that calculates the distances of a particle with a vector of particles
        distances = cdist([atom], voronoiVertices, 'euclidean')

        #find the cavity radius within some precision
        index = np.where( np.abs(distances - cavityT) < 10E-10)

        if(index[1].size > 0):
            
            #Uncomment this block to exclude voids that are positioned outside of the inner solvation shell
            #see if the distance between the voronoi node and the Carbon, is farther than the distance of the last oxigen
            #i.e., the void is "outside" the first solvation shell
            #print('x0Carbon:', x0C,y0C,z0C, '   and maxD:',maxD )
            #if(math.dist([x0C,y0C,z0C], [voronoiVertices[index[1]][0][0],voronoiVertices[index[1]][0][1],voronoiVertices[index[1]][0][2]]) > maxD ):
                #print('********This void is outside 1st solvation shell and will be removed ')         
            #    continue


            #print('++++    vertex',index[1], voronoiVertices[index[1]], ' distance:', distances[0][index[1]][0])
            k=k+1

            verticesVoidsx[k]= voronoiVertices[index[1]][0][0]
            verticesVoidsy[k]= voronoiVertices[index[1]][0][1]
            verticesVoidsz[k]= voronoiVertices[index[1]][0][2]
            cavityVoids[k]=  cavityT - data.particles['Radius'][i] #atomType.type_by_id(atomType[i]).radius



    vunique = set(zip(verticesVoidsx,verticesVoidsy,verticesVoidsz, cavityVoids))
    #print(vunique)

    print(len(vunique)-1)
    #print('# type, x,y,z, cavity')
    
    #add the voids to the ovito list of particles data.particles, add some properties like radius and transparency
    for i,ele in enumerate(vunique):
        if(ele[2]!=0):
            #print('H ',ele[0],ele[1], ele[2],ele[3] )
            data.particles_.create_particle((ele[0],ele[1], ele[2]))
            data.particles_.create_property('Radius')[-1] = ele[3]
            data.particles_.create_property('Transparency')[-1] = 0.4
    print('Time spent on scypy:',time.time()-start)


    writexyz(data, file_outputName, frame)
                #columns =["Particle Type", "Position.X", "Position.Y", "Position.Z", "Radius", "Transparency"])#,
                #frame=frame
                #,multiple_frames=True)
```

```python
def selectCarbonsNNN(frame: int, data):    #these will select the carbons in the first solvation shell

    # This user-defined modifier function gets automatically called by OVITO whenever the data pipeline is newly computed.
    # It receives two arguments from the pipeline system:
    #
    #    frame - The current animation frame number at which the pipeline is being evaluated.
    #    data  - The DataCollection passed in from the pipeline system.
    #            The function may modify the data stored in this DataCollection as needed.
    #



    print('Carbon (name is CKK) id for this file is: ', data.particles['Particle Type'].type_by_name('C').id)
    print('Ctype:', Ctype)

    data.apply(ExpressionSelectionModifier(expression = "ParticleType == "+str(Ctype)+" ")) #for aimd C=3, for md is 2

    #Prefetch the particle property Selection
    selection = data.particles.selection
    #Indices of the selected atoms
    selected_atoms = np.where(selection == 1)[0]

    global x0C, y0C, z0C
    x0C = data.particles.positions[selected_atoms[0]][0]
    y0C = data.particles.positions[selected_atoms[0]][1]
    z0C = data.particles.positions[selected_atoms[0]][2]
    #print('Position of the carbon atom x,y,z:', x0C,y0C,z0C, ' atom selected is: ', selected_atoms)

    NselectedOxygens = 0
    theHatefulEight = np.zeros(8, dtype=int)  #will have the 8 water that represent the first solvation shell
    searchNumber = 16
    global farthest
    farthest = []

    while (NselectedOxygens < 8):
        searchNumber += 1
        NselectedOxygens = 0

        # Set up a neighbor finder for visiting the 12 closest neighbors of each particle.
        finder = ovito.data.NearestNeighborFinder(searchNumber, data)

        # Visit particles closest to some particle (in this case the selected carbon)
        #for neigh in finder.find(selected_atoms):
        for neigh in finder.find_at((x0C,y0C,z0C)):
            if (data.particles.particle_types[neigh.index] == Otype): #5 is for oxygens
                #print(neigh.index, neigh.distance,  data.particles.positions[neigh.index], data.particles.particle_types[neigh.index])
                farthest.append(neigh.distance)
                theHatefulEight[NselectedOxygens]=neigh.index
                NselectedOxygens += 1


    #reset selection and selected the 8 oxygens + SCN
    selection[:]=0

    #Prefetch the particle property Selection
    data.apply(ExpressionSelectionModifier(expression = "(ParticleType == "+str(Stype)+") || (ParticleType =="+str(Ctype)+") || (ParticleType =="+str(Ntype)+")")) #SCN molecule
    selection = data.particles.selection


    selection[theHatefulEight] = 1


    #this method will alter the data.particles.selection directly, and leave selection unaltered
    data.apply(ExpandSelectionModifier(cutoff=1.6))


    #Check the particles that were selected
    #print('after selection from data : ', np.where(data.particles.selection == 1))
    #for item in np.where(data.particles.selection == 1):
    #    print(item,' ', data.particles.particle_types[item])
    #positionHateful = []
    #for hateful in theHatefulEight:
    #    positionHateful = [data.particles[hateful]]

    #cm = np.sum(positionHateful, axis=0)

    #print('oxygens distance from the center of mass')
    #for pos in positionHateful:
    #    print(math.dist(cm,pos))


    #delete the other particles
    data.apply(InvertSelectionModifier())
    data.apply(DeleteSelectedModifier())


    return np.mean(farthest)
```

```python
'''
    can work in two ways:
    1) on the fly, gets the data from computeVoidsScy -> selectCarbonNNN
    2) reads xyz file that already had the generated voids (particlesType == 0 )
'''
def computeVolumes(frame, data):

    #if data.particles != None:
    #    print("There are %i particles with the following properties:" % data.particles.count)
    #    for property_name in data.particles.keys():
    #        print("  '%s'" % property_name)

    ovitoFree = 1

    #how does it know that this is just the undeleted particles?
    data.apply(SelectTypeModifier(property = 'Particle Type', types = {0})) #type zero-> voids just created in selectCarbonsNNN

    #Prefetch the particle property Selection
    selection = data.particles.selection
    #Indices of the selected atoms
    selected_atoms = np.where(selection == 1)[0]

    voidPositions = data.particles.positions[selected_atoms]
    #voidCavity = data.particles['Radius']
    #print(voidPositions)

    voidList = []

    for ele in selected_atoms:
        voidList.append((data.particles.positions[ele][0], data.particles.positions[ele][1],data.particles.positions[ele][2],data.particles['Radius'][ele]))
        #print(voidList[-1])
        #print(ele)

    #do the volumes calculation
    Nsamples = 10000


    #calculates the center of mass of the system
    #will be used as a reference for the reduced cell
    #from which the random point will be sample using Monte Carlo
    cm = np.sum(voidPositions, axis=0)
    if len(voidPositions) == 0:
        cm = [0,0,0]
    else:
        cm /= len(voidPositions)
    #print('CM:', cm)
    
    #maxRange = np.max(voidPositions+np.sign(voidPositions)*voidRadius, axis=0)-np.min(voidPositions+np.sign(voidPositions)*voidRadius, axis=0)
    maxRange = [4,4,4]

    if (len(voidPositions) == 1): #this is the single void case
        maxRange[0] = voidList[-1][3]
        maxRange[1] = voidList[-1][3]
        maxRange[2] = voidList[-1][3]
    #print('Max range:',maxRange)

    #data.particles_.create_particle([0,0,0])
    f = 2.0
    data.particles_.create_particle([cm[0]-f*maxRange[0],cm[1]-f*maxRange[1], cm[2]-f*maxRange[2]])
    data.particles_.create_particle([cm[0]-f*maxRange[0],cm[1]+f*maxRange[1], cm[2]-f*maxRange[2]])
    data.particles_.create_particle([cm[0]+f*maxRange[0],cm[1]-f*maxRange[1], cm[2]-f*maxRange[2]])
    data.particles_.create_particle([cm[0]+f*maxRange[0],cm[1]+f*maxRange[1], cm[2]-f*maxRange[2]])

    data.particles_.create_particle([cm[0]-f*maxRange[0],cm[1]-f*maxRange[1], cm[2]+f*maxRange[2]])
    data.particles_.create_particle([cm[0]-f*maxRange[0],cm[1]+f*maxRange[1], cm[2]+f*maxRange[2]])
    data.particles_.create_particle([cm[0]+f*maxRange[0],cm[1]-f*maxRange[1], cm[2]+f*maxRange[2]])
    data.particles_.create_particle([cm[0]+f*maxRange[0],cm[1]+f*maxRange[1], cm[2]+f*maxRange[2]])


    #these are the reduced cell volume (samples cell)
    #why not 2 max range?
    cellX = 4*maxRange[0]
    cellY = 4*maxRange[1]
    cellZ = 4*maxRange[2]


    countInsideVolume = 0
    countOutsideVolume = 0
    distance = 0
    flag = False
    N = 0
    Nsamples = 100000

    ovitoFree = 1


    #print('Calculating:',frame)
    start = time.time()
    while(N<Nsamples):
        N = N + 1

        #generate a random number in the reducedCell (samples cell), that the monte Carlo will samples
        #if original large cell is used, the monte Carlo will need more points to converge.
        xr = uniform( cm[0]- 2*maxRange[0], cm[0]+2*maxRange[0])
        yr = uniform( cm[1]- 2*maxRange[1], cm[1]+2*maxRange[1])
        zr = uniform( cm[2]- 2*maxRange[2], cm[2]+2*maxRange[2])

        #fsamplesM.write(str(xr)+' '+str(yr)+' '+ str(zr)+'\n')
        flag = False

        for i,ele in enumerate(voidList):
            distance = math.dist((xr,yr,zr), (ele[0],ele[1],ele[2]))


            #!!!!!!!!!
            if  abs(distance) <= (ele[3]*ovitoFree) :    #super care here, ovito free, gives the cavity diameter instead of cavity radius
            #print('\n point inside ',x,y,z,' is inside circle',i,
            #      ' ',(ele[0],ele[1],ele[2]), ' with radius: ',ele[3],
            #     ' dist:',distance)
                flag = True
                break   #ensure that the point is not counted twice, as the spheres can overlap
        if flag == False:
            countOutsideVolume += 1

        else:
            countInsideVolume += 1


    end = time.time()
    #print('Time elapsed:',end-start)
    #print('Points inside:', countInsideVolume)
    #print('Points outside:', countOutsideVolume)

    TotalVolume = cellX*cellY*cellZ
    volumeInside = TotalVolume*(countInsideVolume/(countInsideVolume+countOutsideVolume))
    volumeOutside = TotalVolume*(countOutsideVolume/(countInsideVolume+countOutsideVolume))
    #print('Total volume:', TotalVolume )
    #print('Volume outside:', volumeOutside)
    #print('\nVolume inside:', volumeInside, ' ',100*countInsideVolume/(countInsideVolume+countOutsideVolume),'%')


    
    return (volumeInside, len(voidList))
```

```python
'''
    can work in two ways:
    1) on the fly, gets the data from computeVoidsScy -> selectCarbonNNN
    2) reads xyz file that already had the generated voids (particlesType == 0 )
    
    carefull: Nsamples will determine the precision and time spent. N=1.000.000 is excelente precision
'''
def computeVolumesMean(frame, data):
    #if data.particles != None:
    #    print("There are %i particles with the following properties:" % data.particles.count)
    #    for property_name in data.particles.keys():
    #        print("  '%s'" % property_name)

    ovitoFree = 1

    #how does it know that this is just the undeleted particles?
    data.apply(SelectTypeModifier(property = 'Particle Type', types = {0}))

    #Prefetch the particle property Selection
    selection = data.particles.selection
    #Indices of the selected atoms
    selected_atoms = np.where(selection == 1)[0]

    voidPositions = data.particles.positions[selected_atoms]
    #voidCavity = data.particles['Radius']
    #print(voidPositions)

    voidList = []

    for ele in selected_atoms:
        voidList.append((data.particles.positions[ele][0], data.particles.positions[ele][1],data.particles.positions[ele][2],data.particles['Radius'][ele]))
        #print(voidList[-1])
        #print(ele)


    #calculates the center of mass of the system
    #will be used as a reference for the reduced cell
    #from which the random point will be sample using Monte Carlo
    cm = np.sum(voidPositions, axis=0)
    if len(voidPositions) == 0:
        cm = [0,0,0]
    else:
        cm /= len(voidPositions)
    #print('CM:', cm)

    #maxRange = np.max(voidPositions+np.sign(voidPositions)*voidRadius, axis=0)-np.min(voidPositions+np.sign(voidPositions)*voidRadius, axis=0)
    maxRange = [4,4,4]

    if (len(voidPositions) == 1): #this is the single void case
        maxRange[0] = voidList[-1][3]
        maxRange[1] = voidList[-1][3]
        maxRange[2] = voidList[-1][3]
    #print('Max range:',maxRange)

    #data.particles_.create_particle([0,0,0])
    f = 2.0
    data.particles_.create_particle([cm[0]-f*maxRange[0],cm[1]-f*maxRange[1], cm[2]-f*maxRange[2]])
    data.particles_.create_particle([cm[0]-f*maxRange[0],cm[1]+f*maxRange[1], cm[2]-f*maxRange[2]])
    data.particles_.create_particle([cm[0]+f*maxRange[0],cm[1]-f*maxRange[1], cm[2]-f*maxRange[2]])
    data.particles_.create_particle([cm[0]+f*maxRange[0],cm[1]+f*maxRange[1], cm[2]-f*maxRange[2]])

    data.particles_.create_particle([cm[0]-f*maxRange[0],cm[1]-f*maxRange[1], cm[2]+f*maxRange[2]])
    data.particles_.create_particle([cm[0]-f*maxRange[0],cm[1]+f*maxRange[1], cm[2]+f*maxRange[2]])
    data.particles_.create_particle([cm[0]+f*maxRange[0],cm[1]-f*maxRange[1], cm[2]+f*maxRange[2]])
    data.particles_.create_particle([cm[0]+f*maxRange[0],cm[1]+f*maxRange[1], cm[2]+f*maxRange[2]])


    #these are the reduced cell volume (sample cell) 
    #why not 2 max range?
    cellX = 4*maxRange[0]
    cellY = 4*maxRange[1]
    cellZ = 4*maxRange[2]


    #do the volume calculation, takes the mean of n interations
    volumeTemp = []
    iMeanStd = []
    for itera in range(0,10):
        countInsideVolume = 0
        countOutsideVolume = 0
        distance = 0
        flag = False
        N = 0
        Nsamples = 10000

        ovitoFree = 1


        #print('Calculating:',frame)
        start = time.time()
        while(N<Nsamples):
            N = N + 1

            #generate a random number in the reducedCell (samples cell), that the monte Carlo will samples
            #if original large cell is used, the monte Carlo will need more points to converge.
            xr = uniform( cm[0]- 2*maxRange[0], cm[0]+2*maxRange[0])
            yr = uniform( cm[1]- 2*maxRange[1], cm[1]+2*maxRange[1])
            zr = uniform( cm[2]- 2*maxRange[2], cm[2]+2*maxRange[2])

            #fsamplesM.write(str(xr)+' '+str(yr)+' '+ str(zr)+'\n')
            flag = False

            for i,ele in enumerate(voidList):
                distance = math.dist((xr,yr,zr), (ele[0],ele[1],ele[2]))


                #!!!!!!!!!
                if  abs(distance) <= (ele[3]*ovitoFree) :    #super care here, ovito free, gives the cavity diameter instead of cavity radius
                #print('\n point inside ',x,y,z,' is inside circle',i,
                #      ' ',(ele[0],ele[1],ele[2]), ' with radius: ',ele[3],
                #     ' dist:',distance)
                    flag = True
                    break   #insure that the point is not counted twice, as the sphere can overlap
            if flag == False:
                countOutsideVolume += 1

            else:
                countInsideVolume += 1


        end = time.time()
        #print('Time elapsed:',end-start)
        #print('Points inside:', countInsideVolume)
        #print('Points outside:', countOutsideVolume)

        TotalVolume = cellX*cellY*cellZ
        volumeInside = TotalVolume*(countInsideVolume/(countInsideVolume+countOutsideVolume))
        volumeOutside = TotalVolume*(countOutsideVolume/(countInsideVolume+countOutsideVolume))
        #print('Total volume:', TotalVolume )
        #print('Volume outside:', volumeOutside)
        #print('\nVolume inside:', volumeInside, ' ',100*countInsideVolume/(countInsideVolume+countOutsideVolume),'%')
        volumeTemp.append(volumeInside)

        iMeanStd.append([itera, volumeInside, np.mean(volumeTemp), np.std(volumeTemp)])

    #check what is the volume occupied by the spheres if they were separated
    allSpheresVolume = 0
    for i,ele in enumerate(voidList):
        allSpheresVolume += (4/3)*math.pi*(ovitoFree*ovitoFree*ovitoFree*ele[3]*ele[3]*ele[3])
    #print('Isolated sphere volume: ', allSpheresVolume)


    #analysis

    dfStats = pd.DataFrame(data=iMeanStd, columns=['i','volumeInside','volumeMean','volumeStd'])
    print('Final of '+str(itera)+'MC integration Mean value:',np.mean(volumeTemp), 'with std:',np.std(volumeTemp))

    plt.hist(volumeTemp)
    plt.show()
    #dfStats.plot(x=dfStats['i'], y=dfStats['volumeMean'], kind='line')

    return (np.mean(volumeTemp), len(voidList), dfStats)
    #return (volumeInside, len(voidList))
```

# Calculates the volumes

the volumes are calculated using Monte Carlo integration (see function computeVolumesMean)

Because Monte Carlo is a probabilistic approach, a mean of 9 interations is done.
The precision can also be increased by increasing the number of points.


Important configurations:
___
Nframes :  how many frames of the trajectory will you compute.

```python
#read the MD file


#this is the folder names, in my case i had all this snapshots
MD =['200']#,'220','250','300','320','350'] #[200,220]
AIMD = ['850']#,'1050','1250','1550','2050','2350']

voidVolumesString = ''

#change here for MD, AIMD files
#for frame in AIMD:
for frame in MD: 
    folder = frame
    print('Calculating frame:', frame)
    #file_outputName = 'AIMD/'+folder+'MDvoidsCwithH-VdwH20SCNnone.xyz'
    file_outputName = 'MD/'+folder+'MDvoidsCwithH-FindingSCNVolume.xyz'
    #pipeline = import_file('AIMD/'+folder+'/coord.xyz')
    pipeline = import_file('MD/'+folder+'/coord.xyz')

    print(file_outputName)


    pipeline.add_to_scene()
    data = pipeline.compute()

    ###########
    ## This reads the particle types from the particle names.   Different xyz files have different name of their particles
    Ctype = data.particles['Particle Type'].type_by_name('C').id   #.type_by_name('CKK').id
    Otype = data.particles['Particle Type'].type_by_name('O').id   #.type_by_name('OH2').id
    H1type = data.particles['Particle Type'].type_by_name('H').id  #.type_by_name('H1').id
    #H2type = data.particles['Particle Type'].type_by_name('H1').id  #.type_by_name('H2').id
    H2type = H1type
    Stype = data.particles['Particle Type'].type_by_name('S').id  #.type_by_name('SKK').id
    Ntype = data.particles['Particle Type'].type_by_name('N').id  #.type_by_name('NKK').id
    Natype = data.particles['Particle Type'].type_by_name('Na').id  #.type_by_name('NKK').id
    #####

    #pipeline.modifiers.append(setup_particle_types)   #this sets up the vdW particle radii
    pipeline.modifiers.append(defineVdw)              #setup vdw

    
    #this is necessary for voronoi tesselation to work
    mod = AffineTransformationModifier(
            operate_on = {'cell'}, # Transform box but not the particles.
            transformation = [[1.05, 0, 0, 0],
                              [0, 1.05, 0, 0],
                              [0, 0, 1.05, 0]])
    pipeline.modifiers.append(mod)

    #data.cell_.pbc = (True,True,True)

    # Set up the Voronoi analysis modifier.
    voro = VoronoiAnalysisModifier(
        compute_indices = True,
        use_radii = True,                      #care here!
        edge_threshold = 0.1,
        generate_polyhedra = True
    )
    pipeline.modifiers.append(voro)         #activate voro analysis
    data = pipeline.compute()

    #
    #           compute volumes
    #

    Nframes = pipeline.source.num_frames
    Nframes = 1

    print('Total number of frames:', Nframes)
    voidVolumes = np.empty(Nframes)
    sphereNumber = np.empty(Nframes)
    mean_oxigenDistance = np.empty(Nframes)

    start = time.time()
    print('Calculating volume..')
    for frame in range(0,Nframes,1):
        print('frame:',frame)
        data = pipeline.compute(frame)

        selectCarbonsNNN(frame, data)   #use this one
        calculateVoidsScy(frame, data)  #use this one

        #farthest[frame] = selectCarbonsNNN(frame, data)

        voidVolumes[frame],sphereNumber[frame], dfStats  = computeVolumesMean(frame, data)            #use this one
    print('Time elapsed (s):',time.time() - start)
    
    #this is just to double check the monte carlo reduce cell size 
    #export_file(data, "output.xyz", "xyz",
    #                columns =["Particle Type", "Position.X", "Position.Y", "Position.Z"],
                    #frame=frame,
    #                multiple_frames=True)
    



    #plt.plot(voidVolumes)
    #plt.show()
    #plt.hist(voidVolumes)
    #plt.show()
    #plt.plot(sphereNumber)
    #plt.show()
    #plt.hist(sphereNumber)

    print('Mean volume:',voidVolumes.mean(), ' effective radii is ',0.6203*pow(voidVolumes.mean(),1/3))
    print('standart deviation:',voidVolumes.std())
    print('Mean #sphere', sphereNumber.mean())
    print('Mean #sphere deviation:',sphereNumber.std())
    #print('Mean farthest molecule radii:',mean_oxigenDistance.mean())
    voidVolumesString += f'{folder} {voidVolumes.mean()} {voidVolumes.std()}\n'
print('Voids Summary\n')
print(voidVolumesString)
```

```python
#read the MD file


MD =['200','220','250','300','320','350'] #[200,220]
AIMD = ['850','1050','1250','1550','2050','2350']

#radii = [ [0.1,0.6, 0.25, 0.01, 0.01,0.01],[0.1,0.8, 0.45, 0.01, 0.01,0.01],
#          [0.1,1.0, 0.65, 0.01, 0.01,0.01], [0.1,1.2, 0.8, 0.01, 0.01,0.01],
#          [0.1,1.52, 1.2, 0.01, 0.01,0.01]]
radii = [ [0.01,0.6, 0.01, 0.01, 0.01,0.01],[0.01,0.9, 0.01, 0.01, 0.01,0.01], #oxygen only
          [0.01,1.2, 0.01, 0.01, 0.01,0.01], [0.01,1.3, 0.01, 0.01, 0.01,0.01]]
          #[0.01,1.7, 0.01, 0.01, 0.01,0.01], [0.01,2.0, 0.01, 0.01, 0.01,0.01]]

radii = [ [0.01,0.01, 0.25, 0.01, 0.01,0.01],[0.01,0.01, 0.45, 0.01, 0.01,0.01], #hydrogen only
          [0.01,0.01, 0.65, 0.01, 0.01,0.01], [0.01,0.01, 0.85, 0.01, 0.01,0.01],
          [0.01,0.01, 1.05, 0.01, 0.01,0.01]]

    #vdw = [0.1,1.52, 1.2, 0.01, 0.01,0.01]  #wikipedia vdw   C,O,H, S, N, Na
    #vdw = [1.6,1.52, 1.2, 1.8, 1.55,0.1]  #wikipedia vdw   C,O,H, S, N, Na
    #vdw = [0.77, 0.74,0.46, 1.05,0.74 ,0.1]  #ovito visual values

voidVolumesString = ''
global vectorSize

for radius in radii:
    folder = '850'
    vectorSize = radius
    print('Calculating frame:', vectorSize)
    #file_outputName = 'AIMD/'+folder+'MDvoidsCwithH-VdwH20SCNnone.xyz'
    pipeline = import_file('AIMD/'+folder+'/coord.xyz')

    print(file_outputName)


    pipeline.add_to_scene()
    data = pipeline.compute()

    ###########
    ## This reads the particle types from the particle names.  As different data files have different name of their particles
    Ctype = data.particles['Particle Type'].type_by_name('C').id   #.type_by_name('CKK').id
    Otype = data.particles['Particle Type'].type_by_name('O').id   #.type_by_name('OH2').id
    H1type = data.particles['Particle Type'].type_by_name('H').id  #.type_by_name('H1').id
    #H2type = data.particles['Particle Type'].type_by_name('H1').id  #.type_by_name('H2').id
    H2type = H1type
    Stype = data.particles['Particle Type'].type_by_name('S').id  #.type_by_name('SKK').id
    Ntype = data.particles['Particle Type'].type_by_name('N').id  #.type_by_name('NKK').id
    Natype = data.particles['Particle Type'].type_by_name('Na').id  #.type_by_name('NKK').id
    #####

    #pipeline.modifiers.append(setup_particle_types)   #this sets up the vdW particle radii
    pipeline.modifiers.append(defineVdwVector)              #setup vdw


    mod = AffineTransformationModifier(
            operate_on = {'cell'}, # Transform box but not the particles.
            transformation = [[1.05, 0, 0, 0],
                              [0, 1.05, 0, 0],
                              [0, 0, 1.05, 0]])
    pipeline.modifiers.append(mod)

    #data.cell_.pbc = (True,True,True)

    # Set up the Voronoi analysis modifier.
    voro = VoronoiAnalysisModifier(
        compute_indices = True,
        use_radii = True,                      #care here!
        edge_threshold = 0.1,
        generate_polyhedra = True
    )
    pipeline.modifiers.append(voro)         #activate voro analysis
    data = pipeline.compute()

    #
    #           compute volumes
    #

    Nframes = pipeline.source.num_frames
    Nframes = 1

    print('Total number of frames:', Nframes)
    voidVolumes = np.empty(Nframes)
    sphereNumber = np.empty(Nframes)
    mean_oxigenDistance = np.empty(Nframes)

    start = time.time()
    print('Calculating volume..')
    for frame in range(0,Nframes,1):
        print('frame:',frame)
        data = pipeline.compute(frame)

        selectCarbonsNNN(frame, data)
        calculateVoidsScy(frame, data)


        voidVolumes[frame],sphereNumber[frame], dfStats  = computeVolumesMean(frame, data)            #use this one
    print('Time elapsed:',time.time() - start)


    print('Mean volume:',voidVolumes.mean(), ' effective radii is ',0.6203*pow(voidVolumes.mean(),1/3))
    print('standart deviation:',voidVolumes.std())
    print('Mean #sphere', sphereNumber.mean())
    print('Mean #sphere deviation:',sphereNumber.std())
    #print('Mean farthest molecule radii:',mean_oxigenDistance.mean())
    voidVolumesString += f'{folder} {voidVolumes.mean()} {voidVolumes.std()} {radius[1]} {radius[2]}\n'
print('Voids Summary\n')
print(voidVolumesString)
```
